Q. Membership operation
in, not in are two membership operators and it returns boolean value
print ( '2' in 'Python2.7.8' )
print ( 10 in [ 10 , 10.20 , 10 + 20j , 'Python' ])
print ( 10 in ( 10 , 10.20 , 10 + 20j , 'Python' ))
print ( 2 in { 1 , 2 , 3 })
print ( 3 in { 1 : 100 , 2 : 200 , 3 : 300 })
print ( 10 in range ( 20 ))

```python
# Solution
print ( '2' in 'Python2.7.8' )  # 2 is available in string thus True
print ( 10 in [ 10 , 10.20 , 10 + 20j , 'Python' ]) # 10 is available in list, thus True
print ( 10 in ( 10 , 10.20 , 10 + 20j , 'Python' )) # 10 is available is tuple, Thus True
print ( 2 in { 1 , 2 , 3 }) # 2 is available in set, thus True
print ( 3 in { 1 : 100 , 2 : 200 , 3 : 300 }) # 3 is available in dictionary, Thus True
print ( 10 in range ( 20 )) # 10 is avaiable in range 1,20 thus True
```
