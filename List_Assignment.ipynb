```python
#Write a Python program to find the sum of all elements in a list using loop.
#Input:- [10,20,30,40]
#Output:- 100
#Note: You cannot use list comprehension when result has elemenst less than input list
input_list = [10,20,30,40]
sum = 0
for i in input_list:
    sum += i
print(f'sum of list is : {sum}')
    
```

```python
#Write a Python program to find the largest number from a list using loop.
#Input:- [10,100,2321, 1,200,2]
#Output:- 2321
inp_list = [10,100,2321, 1,200,2]
largest = 0
for i in inp_list:
    if i > largest:
        largest = i        
print(f'Largest of list is : {largest}')

```

```python
#Write a Python program to find the smallest number from a list using loop.
#Input:- [10,100,2321, 1,200,2]
#Output:- 1
import random
inp_list = [10,100,2321, 1,200,2]
smallest = random.choice(inp_list)
for i in inp_list:
    if i < smallest:
        smallest = i        
print(f'Smallest number in list is : {smallest}')

```

```python
#Write a Python program to count the number of strings having length more than 2 and are palindrome in a list using loop.
#Input:- ['ab', 'abc', 'aba', 'xyz', '1991']
#Output:- 2

list_1 = ['ab', 'abc', 'aba', 'xyz', '1991']
count = 0
for i in list_1:
    if i == i[::-1]:
        count+=1
print(count)
    
```
