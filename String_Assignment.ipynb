```python
#Take a string from user and check if it is:-
#     1. alphanumeric
#     2. alphabets
#     3. digit
#     4. all letters are in lower case
#     5. all letters are in upper case
#     6. in title case
#     7. a space character
#     8. numeric
#     9. all number elements in string are decimal

str_inp = input('Enter the alphanumeric string: ')
if str_inp.isalnum() == True:
    print(f"String is alpha numeric:{str_inp}")
else:
    print(f" {str_inp}:string is  not alpha numeric:")
    
#_________________________________________________________________
str_inp = input('Enter the alphabet string: ')
if str_inp.isalpha() == True:
    print(f"String is alpha :{str_inp}")
else:
    print(f" {str_inp}:string is  not alpha:")
#______________________________________________________________________

###Exponents, like ², are also considered to be a digit.
str_inp = input('Enter the digit string: ')
if str_inp.isdigit() == True:
    print(f"String is digit :{str_inp}")
else:
    print(f" {str_inp}:string is  not digit:")
#______________________________________________________
str_inp = input('Enter the lowercase string: ')
if str_inp.islower() == True:
    print(f"String has all alphabets in lowercase :{str_inp}")
else:
    print(f" {str_inp}:string does not have all alphabets in lowercase:")
    
##__________________________________________
str_inp = input('Enter the uppercase string: ')
if str_inp.isupper() == True:
    print(f"String has all alphabets in lowercase :{str_inp}")
else:
    print(f" {str_inp}:string does not have all alphabets in uppercase:")
#________________________________________________________________________________

str_inp = input('Enter the Title string: ')
if str_inp.istitle() == True:
    print(f"String is in Title case :{str_inp}")
else:
    print(f" {str_inp}:String is not in Titlecase:")
#___________________________________________________________________________
str_inp = input('Enter the empty string: ')
if str_inp.isspace() == True:
    print(f"String is empty :{str_inp}")
else:
    print(f" {str_inp}:String is not in empty:")
#_________________________________________________________________________
str_inp = input('Enter the numeric string: ')
if str_inp.isnumeric() == True:
    print(f"String is numeric :{str_inp}")
else:
    print(f" {str_inp}:String is not in numeric:")
    
#________________________________________________________________________
###-This method is used on unicode objects.example- "\u0033" is unicode of 3
str_inp = input('Enter the decimal string: ')
if str_inp.isdecimal() == True:
    print(f"String is decimal :{str_inp}")
else:
    print(f" {str_inp}:String is not decimal:")
```
