Write a function to print the fibonaaci series

```python
# 1st way series starts from zero
def fibonaaci_series():
    f = []
    a = 0
    b = 1
    f.append(a)
    f.append(b)
    while b< 100:
            a,b = b , a+b
            f.append(a+b)
    print(f)
    
```

```python
fibonaaci_series()
```

```python
# 2nd way where series starts with input from user and count of series required

def fibonacci_series():
    n =int(input('Enter the count of fibonacci series required:'))
    num1 = int(input('Enter the initial first number:'))
    num2 = int(input('Enter the initial second number:'))
    fibonacci_list = [num1,num2]
    for i in range(n-2):
        fibonacci_list += [0]
        fibonacci_list[-1] = fibonacci_list[-2] + fibonacci_list[-3]
    print(fibonacci_list)
```

```python
fibonacci_series()
```

What are Armstrong numbers? Write a function if a number is Armstrong number of not
short note on Armstrong number
For example, using a simple number 153 and the decimal system, we see there are 3 digits in it.
If we do a simple mathematical operation of raising each of its digits to the power of 3, and 
then totalling the sum obtained, we get 153. That is 1 to the power of 3 5 to the power of 3 3 to the power 
of three is 1 125 27 153. This can also be represented as 1^3 5^3 3^3=153. 
The number 153 is an example of the Armstrong number which also has a unique property that one can use any number system.     
        

```python
def check_armstrong():
    '''
    short note on Armstrong number For example, using a simple number 153 and the decimal system, we see there are 3 digits in it. 
    If we do a simple mathematical operation of raising each of its digits to the power of 3, and then totalling the sum obtained, we get 153. That is 1 to the power of 3 5 to the power of 3 3 to the power of three is 1 125 27 153. This can also be represented as 1^3 5^3 3^3=153. 
    The number 153 is an example of the Armstrong number which also has a unique property that one can use any number system.
    '''
    num =int(input('Enter the number: '))
    check_digit = len(str(num))
    total = 0
    for i in (str(num)):
        s = int(i)**check_digit
        total += s
    if total == num:
        print(f'{num} is armstrong')
    else:
        print(f'{num} is not armstrong')     
    
    
```

Q5. What are mirrored palindrome? Write a program to check if a string is mirrored palindrome or not?
Given a string S containing only uppercase English characters. 
The task is to find whether S is the same as its reflection in a mirror.

```python

def isReflectionEqual(s):
    # Symmetric characters
    symmetric = dict()
    str1 = "AHIMOTUVWXY" 
    for i in str1:
        symmetric[i] = 1 
    n = len(s)
    for i in range(n):
        if (symmetric[s[i]] == 0):
            return False
       
    rev = s[::-1]
    # Check if the is a palindrome
    if (rev == s):
        return True
    else:
        return False
 
# Code
s = input('Enter the string: ')

if (isReflectionEqual(s)):
    print("YES")
else:
    print("NO")

 
  
```
