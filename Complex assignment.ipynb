print ( 10 + 20j in [ 10 , 10.20 , 10 + 20j , 'Python' ]) #True
print ( 10 + 20j in ( 10 , 10.20 , 10 + 20j , 'Python' )) #True
print ( 30 + 40j in { 1 , 20.30 , 30 + 40j }) #True
print ( 30 + 40j in { 1 : 100 , 2.3 : 200 , 30 + 40j : 300 }) #True
print ( 10 in range ( 20 )) #True
